```
import pandas as pd
starbucks_file = "All_Starbucks_Locations_in_the_US.csv"
```

Funtion to display results.

```
def displayStores(dfStores):
    for index, row in dfStores.iterrows():
        print(row['Name'])
        print(row['Street Address'])
        print(row['City'], row['State'], row['Zip'])
        print(f'\n')
```

Function to read data.

```
def getData():
    fields = ('Store Number', 'Name', 'Street Address', 'City', 'State', 'Zip','Latitude','Longitude','Features - Stations')
    df = pd.read_csv(starbucks_file, skipinitialspace=True, usecols=fields)
    return df
data = getData()
```

Function to create menu.

```
def opening_menu():
    print("*** Starbucks Store Finder ***\n")
    print("1 - Find Stores by City and State\n2 - Find Stores by Zip Code\n3 - Find Stores by Zip Code and with Drive-Through\n4 - Quit")
    opening_menu.user_select = int(input("Enter Choice: "))
    return opening_menu.user_select
```

Function to find locations by City and State.

```
def findStoresByCityState(dfStores):
    print("===== City and State Finder =====")
    city = input("Enter a city: (enter to stop) ").capitalize()
    state = input("Enter a state abbreviation: ").upper()
    store_list = dfStores.loc[(dfStores['City'] == city) & (dfStores['State'] == state)]
    store_list_count = len(store_list.index)
    print(f'\nStarbucks Locations by City and State\nThere is {store_list_count} location(s) satisfying this criteria\n\nName/Location')
    return displayStores(store_list)
```

Function to find stores by zip.

```
def findstoresByZip(dfStores):
    print("===== Zip Code Finder =====")
    zip = input("Enter a zip code: (enter to stop) ")
    store_list = dfStores.loc[dfStores['Zip'].str.startswith(zip, na=False)]
    store_list_count = len(store_list.index)
    print(f'\nStarbucks Locations by Zip\nThere is {store_list_count} location(s) satisfying this criteria\n\nName/Location')
    return displayStores(store_list)
```

Function to find drive-thru stores.

```
def findStoresByDriveThru(dfStores):
  print("===== ZIP Code with Drive-Thru Finder =====")
  zip = input("Enter a zip code: (enter to stop) ")
  store_list = dfStores.loc[(dfStores['Zip'].str.startswith(zip, na=False)) & (dfStores['Features - Stations'] == 'Drive-Through')]
  store_list_count = len(store_list.index)
  print(f'\nStarbucks Locations by Zip\nThere is {store_list_count} location(s) satisfying this criteria\n\nName/Location')
  return displayStores(store_list)
```

User Menu.

```
def pickFunction(user_select):
    if user_select == 1:
        return findStoresByCityState(data)
    elif user_select == 2:
        return findstoresByZip(data)
    elif user_select == 3:
        return findStoresByDriveThru(data)
    elif user_select == 4:
        exit()
```

Main funtion.

```
def main():
    pickFunction(opening_menu())
main()
```
